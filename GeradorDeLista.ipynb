# Gerador de Listas/Dicionários para estudos de bases ##

Gerador de listas ou dicionários para fins de estudos de comportamento de bases.

```python
import pandas as pd
import random
```

```python
#Leitura da planilha com a base de palavras do Dicionário Português:
df = pd.read_excel('dicionario.xlsx')

#Criação de lista com base nas palavras do arquivo lido:
listaPalavras = []

for i in df.index:
    
    listaPalavras.append(df.loc[i,'AC'])
```

```python
def filtroPalavras(qtdLista, qtdCaract, ordem):
    
    '''
    Gera uma lista de palavras aleatórias, com base na quantidade da lista solicitada (qtdLista) e quantidade de
    caracteres desejados (qtdCaract)
    
    qtdCaract >>> Se o parâmetro de qtdCaract for igual a 0, é exibido uma lista aleatória de palavras de diversos caracteres
    
    ordem >>> True ou False, sendo False em ordem alfabética e True inversa
    
    '''
    
    if qtdCaract > 0:
        
        baseInicial = [item for item in listaPalavras if len(str(item)) == qtdCaract]
    
    elif qtdCaract == 0:
        
        baseInicial = [item for item in listaPalavras]

    try:
        resultado = [baseInicial[random.randint(0, len(baseInicial))] for i in range(0, qtdLista)]

        
        if ordem == '':
            
            ordem = False
        
        resultado.sort(reverse = ordem)

    except:

        texto = 'Não foi localizado a quantidade solicitada: {}! Consta {} palavras na base!'.format(qtdLista, len(baseInicial))

        return texto

    return resultado
```

```python
#Um exemplo de geração de base para estudo em forma de dicionário...

#Para facilitar o input ou pode-se incluir diretamente na function(FiltroPalavra) os valores:
qtdLista, qtdCaract, ordem = 10, 8, False

baseEstudos = dict([(item, random.randrange(1,5000)) for item in filtroPalavras(qtdLista, qtdCaract, ordem)])

display(baseEstudos)
```

```python
#Um exemplo de geração de base para estudo em forma de lista com tuplas...

#Para facilitar o input ou pode-se incluir diretamente na function(FiltroPalavra) os valores:
qtdLista, qtdCaract, ordem = 25, 16, False

baseEstudos = [(item, random.randrange(1,5000)) for item in filtroPalavras(qtdLista, qtdCaract, ordem)]

display(baseEstudos)
```

```python
#Um exemplo de geração de base para estudo em forma de lista com palavras de tamanhos aleatórios...

#Para facilitar o input ou pode-se incluir diretamente na function(FiltroPalavra) os valores:
qtdLista, qtdCaract, ordem = 25, 10, False

#Incluído um exemplo de valores aleatórios, apenas para estudos fictícios:
baseEstudos = [(item, random.randrange(1,5000)) for item in filtroPalavras(25,0, False)]

#Criação de um DataFrame no Pandas:
dFrame = pd.DataFrame(baseEstudos, columns = ['Palavra', 'Valor'])

#Visualização do DataFrame:
display(dFrame)
```
